```python
import pypsa
import pandas as pd
import numpy as np
import yaml
import sys
import os
import yaml

sys.path.append(os.path.abspath("../pypsa-eur-resilient/config"))
sys.path.append(os.path.abspath("../pypsa-eur-resilient/scripts"))
```

```python
# open config yaml
RESULTS = os.path.abspath("../pypsa-eur-resilient/results") + "/"
config = yaml.safe_load(open("../pypsa-eur-resilient/config/dev.config.yaml", "r"))
run_name = config["run"]["name"]
columns = config["solve_operations"]["columns"]
planning_horizons = config["scenario"]["planning_horizons"]
clusters = config["scenario"]["clusters"][0]
opts = config["scenario"]["opts"][0]
sector_opts = config["scenario"]["sector_opts"][0]
```

```python
networks = dict()

for planning_horizon in planning_horizons:
    print(f"Loading investment {planning_horizon}...")
    n_inv = pypsa.Network(f"{RESULTS}/{run_name}/networks/base_s_{clusters}_{opts}_{sector_opts}_{planning_horizon}.nc")
    networks[("investment", planning_horizon)] = n_inv

print("")

for column in columns:
    for planning_horizon in planning_horizons:
        print(f"Loading {column} {planning_horizon}...")
        path = RESULTS + run_name + "/networks/" + column + "/"
        file_name = f"base_s_ops_{clusters}_{opts}_{sector_opts}_{planning_horizon}.nc"
        networks[(column, planning_horizon)] = pypsa.Network(path + file_name)

optimisation_runs = sorted({key[0] for key in networks})
```

```python
# Create table of CAPEX
capex = pd.DataFrame(
    index=optimisation_runs,
    columns=planning_horizons,
)

for row in optimisation_runs:
    for planning_horizon in planning_horizons:
        n = networks[(row, planning_horizon)]
        value = n.statistics.capex().sum()
        # get all components
        capex.loc[row, planning_horizon] = value / 1e9 # bn. EUR 

print("### CAPEX in bn. EUR ###")
print(capex)
```

```python
# Create table of emissions
emissions= pd.DataFrame(
    index=optimisation_runs,
    columns=planning_horizons,
)

for row in optimisation_runs:
    for planning_horizon in planning_horizons:
        row_config = config["solve_operations"]["definitions"].get(row, {})

        n_inv = networks[("investment", planning_horizon)]
        n = networks[(row, planning_horizon)]

        if row_config and row_config["options"].get("use_emission_price", False):
            value = -1*(n.generators_t.p.T.loc["co2 atmosphere"]@n.snapshot_weightings.generators)
        else:
            value = 0
        
        emissions.loc[row, planning_horizon] = value / 1e6

print("### Emission costs in Mt ###")
print(emissions.map(lambda x: f"{x:.6f}" if isinstance(x, float) else x))
```

```python
# Create table of emission costs
emission_costs = pd.DataFrame(
    index=optimisation_runs,
    columns=planning_horizons,
)

for row in optimisation_runs:
    for planning_horizon in planning_horizons:
        row_config = config["solve_operations"]["definitions"].get(row, {})

        n_inv = networks[("investment", planning_horizon)]
        n = networks[(row, planning_horizon)]

        if row_config and row_config["options"].get("use_emission_price", False):
            emissions = -1*(n.generators_t.p.T.loc["co2 atmosphere"]@n.snapshot_weightings.generators)
            ep = np.round(abs(n_inv.buses_t.marginal_price.T.loc["co2 atmosphere"].values[0]), 6)
            ep_factor = row_config["options"]["use_emission_price"]*1
            value = ep * ep_factor * emissions
        else:
            value = 0
        
        emission_costs.loc[row, planning_horizon] = value / 1e9

print("### Emission costs in bn. EUR ###")
print(emission_costs)
```

```python
# Create table of OPEX
opex = pd.DataFrame(
    index=optimisation_runs,
    columns=planning_horizons,
)

for row in optimisation_runs:
    for planning_horizon in planning_horizons:
        n = networks[(row, planning_horizon)]
        value = n.statistics.opex().sum()
        # get all components
        opex.loc[row, planning_horizon] = value / 1e9 # bn. EUR 

# Account for emission costs
for row in optimisation_runs:
    for planning_horizon in planning_horizons:
        opex.loc[row, planning_horizon] -= emission_costs.loc[row, planning_horizon]

print("### OPEX in bn. EUR ###")
print(opex.map(lambda x: f"{x:.6f}" if isinstance(x, float) else x))
```
